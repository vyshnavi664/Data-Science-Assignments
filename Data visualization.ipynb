```python
import pandas as pd
df=pd.read_csv('sales_data_with_discounts.csv')
df.head()
```

```python
#Histogram
import matplotlib.pyplot as plt
df['Volume'].hist()
plt.xlabel('Volume')
plt.show()
s=df['Volume'].skew()
print('Skewedness:',s)
if s>0:
    print('Positively skewed and presence of outliers will be on rightside')
elif s==0:
    print('Symetrically skewed and no outliers are present')
else:
    print('Negatively skewed and presence of outliers will be on leftside')
```
