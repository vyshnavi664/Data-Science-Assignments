```python
#loading the data
import pandas as pd
df=pd.read_csv("Cardiotocographic.csv")
df.head()
```

```python
#checking if any nulls 
df.info()
```

```python
#finding statistical summary for each variable in the dataset
df.describe()
```

```python
import matplotlib.pyplot as plt
df.boxplot(column='NSP',vert=False)
plt.show()
```

```python
import seaborn  as sns
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
Identifying Correlations
Strong Positive Correlations: 
"Acceleration" and "Fetal Movements" have a strong positive correlation, it might suggest that increased fetal movements often coincide with increased accelerations in FHR.
Strong Negative Correlations:  "Decelerations Late" and "Fetal Movements" might suggest that fetal movements are associated with a decrease in late decelerations.
Weak or No Correlations: Lighter colors indicate weak or no correlation.
```
